```python
stop = set(stopwords.words('English'))
punctuation = set(string.punctuation)
stopWords = stop.union(punctuation)
stopWords.add('')
stopWords.add(' ')
stopWords
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def posTag(tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_words(tweet):
    words = tweet.split(' ')
    clean_words = []
    for i in words:
        #print(i)
        if i.lower() not in stopWords:
            #print(pos_tag([i]))
            pos = posTag(pos_tag([i])[0][1]) #pos_tag always takes list as input.
            #print(pos)
            clean_word = lemmatizer.lemmatize(i,pos = pos)
            clean_words.append(clean_word)
    return clean_words
```

```python
X = list(X)
X_Train = [clean_words(tweet) for tweet in X]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_Train = [' '.join(tweet) for tweet in X_Train]
```

```python
testData = pd.read_csv(r'C:\Users\venug\Downloads\0000000000002747_test_twitter_x_test.csv')
```

```python
X_OutTest = testData['text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_Train,y,stratify = y)
```

```python
#X_train,X_test,y_train,y_test = train_test_split(X,y,stratify = y)
```

```python
count_vec = CountVectorizer(max_features = 3000) #Accuracy seems to go down with ngrams as features. # Without Lemmatization, achieving similar accuracy requires more number of features and lower regularization. 
a = count_vec.fit_transform(X_train)
a
```

```python
test = count_vec.transform(X_test)
```

```python
svc = SVC(kernel = 'rbf',C = 1000,gamma = 'auto')
svc.fit(a,y_train)
svc.score(test,y_test)
```
